# Datos nulos

```python
df.isnull().any().any()
```

```python
# Ver valores nulos
df.isnull().sum()
```

Debido a que no hay datos nulos, se puede omitir el tratamiento de ellos.

```python
# Eliminar duplicados
df = df.drop_duplicates()
```

```python
# Rellenar nulos (ejemplo)
df = df.fillna(method='ffill')
```

```python
#separacion de las variables categóricas y numéricas

numericas = df.select_dtypes(include=['int64', 'float64']).columns
categoricas = df.select_dtypes(include=['object']).columns

print("Numéricas:", numericas)
print("Categóricas:", categoricas)
```

```python
# Histogramas

cols_importantes = ['transaction_amount', 'credit_score', 'account_age_days']

cols_importantes = [col for col in cols_importantes if col in df.columns]

df[cols_importantes].hist(figsize=(8,6))
plt.show()
```

```python
#analisis de fraude

sns.countplot(x='label', data=df)
plt.show()
```

```python
#Correlacion
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, cmap='coolwarm', linewidth=.5)
plt.title("Mapa de correlación general")
plt.show()
```

Se observa que algunas variables tienen correlación positiva con el fraude, lo que indica que podrían ser relevantes para un modelo predictivo. Sin embargo, otras variables no muestran relación significativa.

```python
posibles_targets = ['label', 'is_fraud', 'fraud', 'target']

target = None
for col in posibles_targets:
    if col in df.columns:
        target = col
        break

print("Variable objetivo detectada:", target)
```

```python
if 'transaction_amount' in df.columns:
    sns.boxplot(x=target, y='transaction_amount', data=df)
    plt.show()
else:
    print("No existe la columna transaction_amount")
```
